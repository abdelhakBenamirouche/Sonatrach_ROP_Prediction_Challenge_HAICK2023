### Plot correlation map

```python
def plot_correlation_map(df):
    corr = df.corr()
    s ,ax = plt.subplots(figsize = (14,12))
    cmap = sns.diverging_palette(250, 100, s=75, l=40, n=9, as_cmap = True)
    s = sns.heatmap(corr, cmap=cmap, square=True, cbar_kws={"shrink":.9}, ax=ax, annot=True, annot_kws={'fontsize':12})
```

```python
plot_correlation_map(data)
```

##  Spliting your dataset into a training set and a testing set

```python
from sklearn.model_selection import train_test_split
```

```python
selected_features = ["Torque Avg","Torque Max","Torque Min","RPM Avg","RPM Inst","WOB Max",
                     "WOB Avg","Flow Paddle","Flow In Pumps","Pit 6 Volume","Pressure: SPP Avg"]
```

```python
# Extract X and y from our data

X = data.drop("ROP (Time)", axis = 1).values
y = data["ROP (Time)"].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 20)
```

## XGboost regressor

```python
import xgboost
```

```python
from sklearn import metrics
from sklearn.metrics import mean_squared_error
```

```python
xgb = xgboost.XGBRegressor(n_estimators=100000, max_depth=20, eta=0.001, subsample=0.7, colsample_bytree=0.8)
xgb.fit(X_train, y_train)
predicted = xgb.predict(X_test)
```

```python
print("RMSE", np.sqrt(mean_squared_error(y_test,predicted)))
print("R squared", metrics.r2_score(y_test,predicted))
```

```python
test_data = pd.read_csv("test_dataset.csv")
test_data.head()
```

```python
predicted = xgb.predict(test_data)
pred = np.array([i for i in predicted])
```

```python
sub = pd.read_csv("sample.csv")
sub.head()
```

```python
sub["ROP (Time)"] = pred
sub.set_index('ID', inplace = True)
```

```python
sub.head()
```

```python
sub.to_csv("sonatrach_8th_sub.csv")
```

## Fine tune Xgboost

```python
import xgboost as xgb
 
dmatrix = xgb.DMatrix(data=X, label=y)
params = {'objective':'reg:squarederror'}
cv_results = xgb.cv(dtrain=dmatrix, 
                    params=params,
                    nfold=10, 
                    metrics={'rmse'})
print('RMSE: %.2f' % cv_results['test-rmse-mean'].min())
```

```python
import xgboost as xgb

xgbr = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=20000,
                        max_depth=10, eta=0.001, subsample=0.8, colsample_bytree=0.8)
xgbr.fit(X_train, y_train)
 
ypred = xgbr.predict(X_test)
mse = mean_squared_error(y_test, ypred)
print("RMSE: %.2f" % (mse**(1/2.0)))
```
